```python
import pandas as pd
from pyspark.sql import functions as f
from pyspark.sql.window import Window
c = f.col
from calendar import monthrange
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
from pyspark.sql.types import *
```

```python
df_books = spark.read.csv('galbraith.csv', header=True)
df_books.show()
```

```python
df_books
```

```python
df_books.toPandas()
```

```python
df_books_sums = df_books.select(['book_name', 
                 sum([c(x).cast('double') for x in df_books.columns[1:700]]).alias('sum1'),
                 sum([c(x).cast('double') for x in df_books.columns[700:1400]]).alias('sum2'),
                 sum([c(x).cast('double') for x in df_books.columns[1400:2100]]).alias('sum3'),
                 sum([c(x).cast('double') for x in df_books.columns[2100:2800]]).alias('sum4'),
                 sum([c(x).cast('double') for x in df_books.columns[2800:]]).alias('sum5')]).cache()
df_books_sums.show(50)
```

```python
df_books_sum = df_books_sums.select(['book_name', sum([c(x) for x in df_books_sums.columns[1:]]).alias('sum')])
df_books_sum.show(50)
```
